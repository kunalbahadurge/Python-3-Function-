    6.Explain the concept of generators in Python and how they are defined ?
    = Generators are special functions that return values one at a time instead of all at once. They use the yield keyword instead of return. Generators are memory-efficient and pause after each yield until the next value is requested.

    Example:

```python
def numbers():
    yield 1
    yield 2
    yield 3

for n in numbers():
    print(n)
```

    7.What are the advantages of using generators over regular functions ?
    = Generators use less memory because they produce values one at a time instead of storing everything.They are faster for large data and provide lazy evaluation, meaning they generate a value only when needed.

    Example:

```python
squares = (x * x for x in range(5))
print(next(squares))
print(next(squares))
```

    10.What is the difference between map(), reduce(), and filter() functions in Python?
    
    = map() → Applies a function to every item and returns the transformed items.
    filter() → Keeps only the items that satisfy a condition.
    reduce() → Repeatedly applies a function to all items and reduces them to a single value.
    
    Example:

```python
from functools import reduce

nums = [1, 2, 3, 4]

m = map(lambda x: x * 2, nums)
f = filter(lambda x: x % 2 == 0, nums)
r = reduce(lambda a, b: a + b, nums)

print(list(m))
print(list(f))
print(r)
```

    1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_even(nums):
    total = 0
    for n in nums:
        if n % 2 == 0:
            total += n
    return total

print(sum_even([1, 2, 3, 4, 6, 9]))
```

    4.Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(num)
```

    5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        if self.count == 1:
            self.count += 1
            return self.b
        next_num = self.a + self.b
        self.a = self.b
        self.b = next_num
        self.count += 1
        return next_num

for n in Fibonacci(10):
    print(n)
```

    6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two(limit):
    for e in range(limit + 1):
        yield 2 ** e

for x in power_of_two(5):
    print(x)
```

    7.Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_lines(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()

for line in read_file_lines("/content/sample_data/README.md"):
    print(line)
```

    11.Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €.

    Write a Python program using lambda and map

```python
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

result = list(map(lambda order:
                  (order[0],
                   order[2] * order[3] if order[2] * order[3] >= 100
                   else order[2] * order[3] + 10),
                  orders))

print(result)
```
